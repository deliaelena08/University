```python
from scipy.stats import norm
import numpy as np
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot,title,xlabel,ylabel
#generati zece mii de valori pt x
m=165
sigma=20
prob_inalt=1-norm.cdf(190,loc=m,scale=sigma)
N=10000
nr_selectii=[]
for _ in range(N):
    numar_persoane=0
    while True:
        inaltime=np.random.normal(m,sigma)
        numar_persoane+=1
        if inaltime>190:
            break
    nr_selectii.append(numar_persoane)
#hitograma frecventelor relative pentru valorile 1,2,..10
hist(nr_selectii,range=[1,10],density=True, alpha=0.6, color='bisque', edgecolor='black')
title('Histograma frecventelor relative pentru numarul de selectii (X)')
xlabel('Numarul de selectii (X)')
ylabel('Frecventa relativa')
grid(True)
show()
```

```python
#estimati P(x>10) si afisati probabilitatea teoretica
prob_sim = sum(1 for x in nr_selectii if x > 10) / N
print(f"Probabilitatea estimata P(X > 10) este: {prob_sim}")

# Distribuția X urmează o distribuție geometrică cu probabilitatea de succes prob_persoana_inalta
prob_teoretic = (1 - prob_inalt) ** 10
print(f"Probabilitatea teoretică P(X > 10) este: {prob_teoretic}")
```
