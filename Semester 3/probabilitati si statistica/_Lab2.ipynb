```python
from matplotlib.pyplot import plot, grid, title, show
from random import randint

lista = []
flag = 0
prob = 0
prob_lista = []
for n in range(2,50):
    for i in range(1000):
        for j in range(n):
            lista.append(randint(1,365))
        for a in range(n-1):
            for b in range(a+1,n):
                if lista[a] == lista[b]:
                    flag = 1
        if(flag):
              prob += 1
        flag = 0
        lista.clear()
    prob_lista.append(prob/10)
    prob = 0

title('Plot test')
xs = range(len(prob_lista))
ys = prob_lista
plot(xs,ys,'r*')
grid()
show()
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist

lista = [500,1000, 2000]
prob_i = 0
for N in lista:
  axis('square')
  axis((0, 1, 0, 1))
  for i in range(N):
    F=[random(),random()]
    if dist(F,(0.5,0.5)) < 0.5:
      prob_i += 1
      plot(F[0],F[1],'ro')
    else:
      plot(F[0],F[1],'go')
  print(prob_i/N*4)
  show()
  prob_i=0
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist

lista = [500,1000, 2000]
prob_ii = 0
for N in lista:
  axis('square')
  axis((0, 1, 0, 1))
  for i in range(N):
    F=[random(),random()]
    if(F[0]<=0.5 and F[1] <= 0.5):
      if dist(F,(0.5,0.5)) < dist(F,(0,0)):
        prob_ii += 1
        plot(F[0],F[1],'ro')
    elif(F[0]<=0.5 and F[1] >= 0.5):
      if dist(F,(0.5,0.5)) < dist(F,(0,1)):
        prob_ii += 1
        plot(F[0],F[1],'ro')
    elif(F[0]>=0.5 and F[1] >= 0.5):
      if dist(F,(0.5,0.5)) < dist(F,(1,1)):
        prob_ii += 1
        plot(F[0],F[1],'ro')
    elif(F[0]>=0.5 and F[1] <= 0.5):
      if dist(F,(0.5,0.5)) < dist(F,(1,0)):
        prob_ii += 1
        plot(F[0],F[1],'ro')
  print(prob_ii/N)
  show()
  prob_ii=0
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist

lista = [500,1000, 2000]
prob_iii = 0
for N in lista:
  axis('square')
  axis((0, 1, 0, 1))
  for i in range(N):
    F=[random(),random()]
    dist1 = dist(F,(0,0))
    dist2=dist(F,(0,1))
    dist3= dist(F,(1,1))
    dist4=dist(F,(1,0))
    v = [dist1**2 + dist2**2 < 1, dist2**2 + dist3**2 < 1, dist3**2 + dist4**2 < 1, dist4**2 + dist1**2 < 1]
    if(sum(v) == 2):
      prob_iii += 1
      plot(F[0],F[1],'ro')
    else:
      plot(F[0],F[1],'go')
  print(prob_iii/N)
  show()
  prob_iii=0
```
