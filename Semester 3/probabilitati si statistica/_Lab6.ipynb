```python
from scipy.stats import norm
import random
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot
# --------1.i--------------
n=random.choice([1000,2000,5000])
m=165
sigma=10
samples = norm.rvs(loc=m, scale=sigma, size=n)

hist(samples, bins=14, range=[130, 200], density=True, alpha=0.6, color='bisque', edgecolor='black', label='Relative frequencies')

x = linspace(130, 200, 500)  
pdf = norm.pdf(x, loc=m, scale=sigma)  
plot(x, pdf, 'm-', linewidth=2, label='Functia densitate')
grid(True)
legend()
xticks(range(130, 201, 10))  
show()


#-------1.ii------------

mean_value = mean(samples)
#mean_value= sum([160<x<170 for x in samples])/len(samples)
std_dev = std(samples)
proportion_in_range = ((samples >= 160) & (samples <= 170)).sum() / n

exact_mean = m
exact_std = sigma
exact_proportion = norm.cdf(170, loc=m, scale=sigma) - norm.cdf(160, loc=m, scale=sigma)

print(f"Valoarea medie: {mean_value:.2f}, Valoarea exacta: {exact_mean:.2f}")
print(f"Derivatia standard: {std_dev:.2f}, Derivatia exacta: {exact_std:.2f}")
print(f"Proportia de valori [160, 170]: {proportion_in_range:.4f}, Proportia exacta [160, 170]: {exact_proportion:.4f}")

```
