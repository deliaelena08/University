```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
#a
X=norm.rvs(100,20,1000)
prob=sum([1 for k in X if (k<140 and k>80)])/len(X)
print("Probabilitatea este: "+str(prob))
#b
probt = norm.cdf(140, loc=100, scale=20) - norm.cdf(80, loc=100, scale=20)
print(f"Probabilitatea teoretică P(80 <= X < 140) este: {probt}")
#c
plt.hist(X, bins=13, density=True, alpha=0.6, color='bisque', edgecolor='black', rwidth=0.9, label='Frecvențe relative')
x = np.linspace(min(X), max(X), 500)
pdf = norm.pdf(x, loc=100, scale=20)
plt.plot(x, pdf, 'm-', linewidth=2, label='Funcția de densitate teoretică')
plt.title('Histograma frecvențelor relative și funcția de densitate teoretică')
plt.xlabel('Valori ale pulsului (X)')
plt.ylabel('Densitate de probabilitate')
plt.legend()
plt.grid(True)
plt.show()
```
